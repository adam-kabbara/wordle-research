```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    df_word2vec = pd.read_csv('data/word2vec_distance_statistics.csv')
    df_glove = pd.read_csv('data/glove_distance_statistics.csv')
except FileNotFoundError:
    os.chdir("../")
    print(f"Changed dir to {os.getcwd()}")
    df_word2vec = pd.read_csv('data/word2vec_distance_statistics.csv')
    df_glove = pd.read_csv('data/glove_distance_statistics.csv')
    print("Loaded data from 'data/word2vec_distance_statistics.csv'")
```

```python
df_green_word2vec = df_word2vec.iloc[:, :2].sort_values(by='state', ascending=True, key=lambda col: col.str[0])
df_green_word2vec['green_num'] = df_green_word2vec['state'].str[0]

df_green_glove = df_glove.iloc[:, :2].sort_values(by='state', ascending=True, key=lambda col: col.str[0])
df_green_glove['green_num'] = df_green_glove['state'].str[0]
df_green_glove
```

```python
sns.violinplot(
    data=df_green_glove,
    y='green_num',
    x='Cohen Distance',
    orient='h',
)

plt.title('Violin Plot of Cohen d by Green Squares Count (GloVe Distance)')
plt.xlabel('Cohen d')
plt.ylabel('Number of Green Squares')
plt.tight_layout()
# save to pdf
plt.savefig('violin_plot_green_glove.pdf')
plt.show()

sns.violinplot(
    data=df_green_word2vec,
    y='green_num',
    x='Cohen Distance',
    orient='h',
)
plt.title('Violin Plot of Cohen d by Green Squares Count (Word2Vec Distance)')
plt.xlabel('Cohen d')
plt.ylabel('Number of Green Squares')
plt.tight_layout()
# save to pdf
plt.savefig('violin_plot_green_word2vec.pdf')
plt.show()
```

```python
sns.boxplot(
    data=df_grey_glove,
    y='grey_num',
    x='Cohen Distance',
    orient='h'
)
sns.stripplot(
    data=df_grey_glove,
    y='grey_num',
    x='Cohen Distance',
    orient='h',
    color='black',
    size=3,
    jitter=True,
    alpha=0.6
)
plt.title('Box and Strip Plot of Cohen d by Grey Squares Count (GloVe Distance)')
plt.xlabel('Cohen d')
plt.ylabel('Number of Grey Squares')
plt.tight_layout()
# save to pdf
plt.savefig('box_strip_plot_grey_glove.pdf')
plt.show()

sns.boxplot(
    data=df_grey_word2vec,
    y='grey_num',
    x='Cohen Distance',
    orient='h'
)

sns.stripplot(
    data=df_grey_word2vec,
    y='grey_num',
    x='Cohen Distance',
    orient='h',
    color='black',
    size=3,
    jitter=True,
    alpha=0.6
)

plt.title('Box and Strip Plot of Cohen d by Grey Squares Count (Word2Vec Distance)')
plt.xlabel('Cohen d')
plt.ylabel('Number of Grey Squares')
plt.tight_layout()
# save to pdf
plt.savefig('box_strip_plot_grey_word2vec.pdf')
plt.show()
```
